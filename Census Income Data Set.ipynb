```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
cols = ['age', 'work-class', 'final-weight', 'education', 'education-num', 'marital-status',
       'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week',
       'native-country', 'income']
```

```python
df = pd.read_csv("adult.csv", names=cols)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns = [i.strip()for i in df.columns]
```

```python
for i in df.columns:
    print(df[i].unique())
    print("="*64)
```

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df["income"] = df["income"].map({"<=50K": 1, ">50K": 2})
```

```python
df.head()
```

```python
df.info()
```

# Data distribution

```python
numerical = df.columns[df.dtypes.values != "object"]
```

```python
plt.figure(figsize=(17, 15))
plt.suptitle('Univariate Analysis of Numerical Features', 
             fontsize=20, 
             fontweight='bold', 
             alpha=0.8, 
             y=1.)

for i in range(0, len(numerical)-1):
    plt.subplot(3, 3, i+1)
    ax = sns.kdeplot(df[numerical[i]], shade=True, color='r')
    plt.xlabel(numerical[i], fontsize=16)
    plt.xticks(rotation=90, fontsize=16)
    ax.set_ylabel('Density', fontsize=16)
    plt.yticks(fontsize=16)
    plt.tight_layout()
```

# Count Plot

```python
categorical = df.columns[df.dtypes.values == "object"]
```

```python
plt.figure(figsize=(20, 20))
plt.suptitle('Counts of Categorical Features', 
             fontsize=20, 
             fontweight='bold', 
             alpha=0.8, 
             y=1.)

for i in range(0, len(categorical)-3):
    plt.subplot(4, 2, i+1)
    ax = sns.countplot(x=df[categorical[i]], order=df[categorical[i]].value_counts().index, palette="Set2") # Palette is color type
    plt.xlabel(categorical[i], fontsize=16)
    # Use to rotate the x tick labels
    plt.xticks(rotation=90, fontsize=16) # We can use 90 deg to change the x tick labels to vertical 
    ax.set_ylabel('Count', fontsize=16)
    plt.yticks(fontsize=16)
    plt.tight_layout() # use to align the charts closer
```

```python
categorical_3 = ['sex', 'income', "native-country"]
```

```python
plt.figure(figsize=(20, 25))
plt.suptitle('Counts of Categorical Features', 
             fontsize=20, 
             fontweight='bold', 
             alpha=0.8, 
             y=1.)

for i in range(0, len(categorical_3)):
    plt.subplot(3, 2, i+1)
    ax = sns.countplot(x=df[categorical_3[i]], order=df[categorical_3[i]].value_counts().index, palette="Set2") # Palette is color type
    plt.xlabel(categorical_3[i], fontsize=16)
    # Use to rotate the x tick labels
    plt.xticks(rotation=90, fontsize=16) # We can use 90 deg to change the x tick labels to vertical 
    ax.set_ylabel('Count', fontsize=16)
    plt.yticks(fontsize=16)
    plt.tight_layout() # use to align the charts closer
```
